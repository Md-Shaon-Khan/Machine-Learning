```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN,KMeans,AgglomerativeClustering
```

```python
X,y_true = make_blobs(n_samples=500,centers=3,cluster_std=1.3,random_state=42)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0],X[:,1],c='grey',edgecolors='k')
plt.title('Data before clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
dbscan = DBSCAN(eps=1.2,min_samples=8)
label_dbscan = dbscan.fit_predict(X)
```

```python
set(label_dbscan)
```

```python
plt.figure(figsize=(6,5))
unique_labels = set(label_dbscan)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 10, len(unique_labels))]
```

```python
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1] 
    class_member_mask = (label_dbscan == k)
    xy = X[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=[col], label=f"Cluster {k}" if k != -1 else "Noise", edgecolor='k')

plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
X,y_true = make_blobs(n_samples=500,centers=3,cluster_std=2.9,random_state=42)


kmeans = KMeans(n_clusters=3, random_state=42)
labels_kmeans = kmeans.fit_predict(X)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=labels_kmeans, cmap='viridis', edgecolor='k')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
hier = AgglomerativeClustering(n_clusters=3)
labels_hier = hier.fit_predict(X)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=labels_hier, cmap='coolwarm', edgecolor='k')
plt.title("Hierarchical Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np

# Create a concentric circle dataset
X, _ = make_circles(n_samples=5500, factor=.5, noise=.01, random_state=4)

# Apply DBSCAN to the dataset
dbscan = DBSCAN(eps=.01, min_samples=10)
clusters = dbscan.fit_predict(X)

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Concentric Circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
