```python
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = iris.data
y = iris.target


X_scaled = StandardScaler().fit_transform(X)
```

```python
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

print("Original shape:", X_scaled.shape)
print("Reduced shape:", X_tsne.shape)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, cmap='coolwarm', s=50, edgecolor='k')
plt.title("Iris Dataset Visualized using t-SNE")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(label="Class")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='coolwarm', s=50, edgecolor='k')
plt.title("PCA Projection (Linear)")

plt.subplot(1,2,2)
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, cmap='coolwarm', s=50, edgecolor='k')
plt.title("t-SNE Projection (Nonlinear)")

plt.show()
```
