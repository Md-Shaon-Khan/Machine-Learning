```python
import pandas as pd
```

```python
data = {'Category':['A','B','C','D','B','D','A','A']}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
one_hot_encoded = pd.get_dummies(df,columns=['Category'])
one_hot_encoded
```

```python
one_hot_encoded = pd.get_dummies(df, columns=['Category'],prefix='C')
one_hot_encoded
```

```python
one_hot_encoded = pd.get_dummies(df, columns=['Category'], prefix='Dummy',drop_first=True )
one_hot_encoded
```
