```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("nasdaq100.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Date'])
```

```python
df.head()
```

```python
plt.scatter(df['Starting'],df['Ending'])
plt.xlabel("Starting")
plt.ylabel("Ending")
plt.title("Nasaq100 Stock Prices")
```

```python
x = df.drop('Ending',axis=1)
x.head()
```

```python
y = df[['Ending']]
y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x.mean()
```

```python
x.mean()
```

```python
plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Starting'],df['Ending'])
plt.xlabel("Starting")
plt.ylabel("Ending")
plt.title("Nasaq100 Stock Prices")
```

```python
reg.fit(x,y)
```

```python
m = reg.coef_
m
```

```python
c = reg.intercept_
c
```

```python
m*16700 + c
```

```python
reg.predict([[16700]])
```

```python
df['Predicted_y'] = reg.predict(x)
df.head()
```

```python
plt.plot(x,reg.predict(x))
plt.scatter(df['Starting'],df['Ending'])
plt.xlabel("Starting")
plt.ylabel("Ending")
plt.title("Nasaq100 Stock Prices")
```
